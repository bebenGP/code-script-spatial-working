```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
folder_data = r"C:\Users\bputra\OneDrive - ESRI Indonesia\Data Services\data-item-attachment-nationalhub-arcgishub"
file_data = f"{folder_data}\\tracking_data_progress_scrapping.csv"
print(file_data)
```

```python
data = pd.read_csv(file_data, delimiter=';')
df = pd.DataFrame(data)
df
```

```python
field = 'BRAND'
all_text = ' '.join(data[field].dropna())

# Membuat WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Tampilkan WordCloud menggunakan Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from collections import Counter

# Menghitung frekuensi kata-kata dalam teks
word_freq = Counter(all_text.split())

# Menampilkan kata-kata yang paling banyak muncul
most_common_words = word_freq.most_common(50)  # Mengambil 20 kata teratas
#print(most_common_words)

for word, freq in most_common_words:
    print(f"{word}: {freq}")
```
