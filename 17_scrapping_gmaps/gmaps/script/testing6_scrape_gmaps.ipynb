```python
import requests
from bs4 import BeautifulSoup
import csv

output_save_data = r'C:\Users\bputra\OneDrive - ESRI Indonesia\PT ESRI Indonesia\Research\code-script\17_scrapping_gmaps\gmaps\output\testing1_scrape_google_maps.csv'
# Definisikan fieldnames secara global
fieldnames = ['href', 'poi_name', 'rating', 'address', 'tags']

def scrape_data_to_csv(soup):
    tes = soup.find_all('div', class_='bJzME tTVLSc')

    with open( output_save_data, 'a', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames, delimiter=';')

        # Menulis header jika file kosong
        if csvfile.tell() == 0:
            writer.writeheader()

        for cari in tes:
            href = cari.find('a', class_='hfpxzc')
            href_value = href.get('href') if href else None
            
            poi_names = cari.find('div', class_='qBF1Pd fontHeadlineSmall')
            data_poi_name = poi_names.text.strip() if poi_names else None
            
            rating = cari.find('span', class_='MW4etd')
            ratings = rating.text.strip() if rating else None
            
            alamat_elem = cari.find_all('div', class_='W4Efsd')[1].text.strip()
            tags_elem = cari.find_all('div', class_='W4Efsd')[2].text.strip()
            
            writer.writerow({'href': href_value, 'poi_name': data_poi_name, 'rating': ratings, 'address': alamat_elem, 'tags': tags_elem})
            
    print("Data telah disimpan ke dalam file data_tempat.csv")

def scrape_google_maps_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    scrape_data_to_csv(soup)

urls = [
    "https://www.google.com/maps/search/Bulutangkis/@-0.8995325,100.4379122,12z?entry=ttu",
    "https://www.google.com/maps/search/Bulutangkis/@-6.2725674,106.810043,12z?entry=ttu",
    "https://www.google.com/maps/search/Bulutangkis/@-6.5934944,106.7993574,12z?entry=ttu"
]

for url in urls:
    #print("Scraping data dari:", url)
    scrape_google_maps_data(url)
    #print("Selesai scraping dari:", url)
    print("Scrape")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd

options = Options()
options.headless = True
driver = webdriver.Chrome(options=options)

driver.get(url)

# Dapatkan HTML halaman web setelah di-render
html = driver.page_source

# Gunakan BeautifulSoup untuk mengekstrak data
soup = BeautifulSoup(html, 'lxml')

tes = soup.find_all('div', class_='bJzME tTVLSc')

data = []

for cari in tes:
    hrefs = cari.find_all('a', class_='hfpxzc')
    for href in hrefs:
        href_value = href.get('href')
                
    poi_names = cari.find_all('div', class_='qBF1Pd fontHeadlineSmall')
    for poi in poi_names:
        poinames_value = poi.text.strip() 
    
    ratings = cari.find_all('span', class_='MW4etd')
    for rating in ratings:
        ratings_value = rating.text.strip()
                    
    tagsaddress_elements = cari.find_all('div', class_='W4Efsd')
    simpan =[]
    for tagsaddress in tagsaddress_elements:
        tagsaddress_value = tagsaddress.text.strip()
        simpan.append(tagsaddress_value)
        
    data.append([href_value, poinames_value, ratings_value, ';'.join(simpan)])

driver.quit()

# Membuat DataFrame
df = pd.DataFrame(data, columns=['Href', 'Poi Name', 'Rating', 'Address'])

# Menyimpan DataFrame ke dalam file CSV
df.to_csv('data2.csv', sep=';', index=False)
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

options = Options()
options.headless = True
driver = webdriver.Chrome(options=options)

driver.get(url)

# Dapatkan HTML halaman web setelah di-render
html = driver.page_source

# Gunakan BeautifulSoup untuk mengekstrak data
soup = BeautifulSoup(html, 'lxml')

tes = soup.find_all('div', class_='bJzME tTVLSc')

# Buat file CSV untuk menyimpan data
with open('data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Href', 'Poi Name', 'Rating', 'Address']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames, delimiter=';')
    
    writer.writeheader()

    for cari in tes:
        hrefs = cari.find_all('a', class_='hfpxzc')
        poi_names = cari.find_all('div', class_='qBF1Pd fontHeadlineSmall')
        ratings = cari.find_all('span', class_='MW4etd')
        tagsaddress_elements = cari.find_all('div', class_='W4Efsd')

        # Memastikan setiap elemen yang dicari ada sebelum menambahkannya ke dalam CSV
        if hrefs and poi_names and ratings and tagsaddress_elements:
            for href, poi, rating, tagsaddress in zip(hrefs, poi_names, ratings, tagsaddress_elements):
                href_value = href.get('href')
                poinames_value = poi.text.strip()
                ratings_value = rating.text.strip()
                tagsaddress_value = tagsaddress.text.strip()
                print(tagsaddress_value)

                # Tulis data ke dalam file CSV
                writer.writerow({'Href': href_value, 'Poi Name': poinames_value, 'Rating': ratings_value, 'Address': tagsaddress_value})

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

options = Options()
options.headless = True
driver = webdriver.Chrome(options=options)

def scrape_data(url):
    driver.get(url)

    # Dapatkan HTML halaman web setelah di-render
    html = driver.page_source

    # Gunakan BeautifulSoup untuk mengekstrak data
    soup = BeautifulSoup(html, 'html.parser')

    tes = soup.find_all('div', class_='bJzME tTVLSc')
    for cari in tes:
        href = cari.find('a', class_='hfpxzc')
        href_value = href.get('href') if href else None
        print("href:", href_value)
        
        poi_names = cari.find('div', class_='qBF1Pd fontHeadlineSmall')
        data_poi_name = poi_names.text.strip() if poi_names else None
        print("poi_name:", data_poi_name)
        
        rating = cari.find('span', class_='MW4etd')
        ratings = rating.text.strip() if rating else None
        print("rating:", ratings)
        
        # Mencari alamat dan tags dengan mengakses elemen kedua dan ketiga dengan class 'W4Efsd'
        alamat_elem = cari.find_all('div', class_='W4Efsd')[1].text.strip()
        if alamat_elem:
            print("address:", alamat_elem)
        
        tags_elem = cari.find_all('div', class_='W4Efsd')[2].text.strip()
        if tags_elem:
            print("tags:", tags_elem)
            
        print()  # Untuk memisahkan setiap hasil

# List URL yang ingin Anda scrape
urls = [
    "https://www.google.com/maps/search/Bulutangkis/@-0.8995325,100.4379122,12z?entry=ttu",
    "https://www.google.com/maps/search/Bulutangkis/@-6.2725674,106.810043,12z?entry=ttu",
    "https://www.google.com/maps/search/Bulutangkis/@-6.5934944,106.7993574,12z?entry=ttu"
]

for url in urls:
    print("Scraping data dari:", url)
    scrape_data(url)
    print("Selesai scraping dari:", url)

driver.quit()
```
