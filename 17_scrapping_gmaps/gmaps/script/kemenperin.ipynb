### Contoh 

```python
from selenium import webdriver
import csv

# Inisialisasi WebDriver (gunakan WebDriver sesuai dengan browser yang Anda gunakan)
driver = webdriver.Chrome()

# Buka halaman web
driver.get("URL_WEBSITE_ANDA")

try:
    # Temukan elemen dengan CSS yang diberikan
    element = driver.find_element_by_css_selector("#newspaper-a > tbody:nth-child(2) > tr:nth-child(1) > td:nth-child(2) > b:nth-child(1)")

    # Dapatkan teks dari elemen
    text_value = element.text

    # Tulis teks ke dalam file CSV
    with open('output.csv', 'w', newline='') as csvfile:
        fieldnames = ['Element_Text']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        # Tulis header
        writer.writeheader()

        # Tulis data
        writer.writerow({'Element_Text': text_value})

finally:
    # Tutup browser
    driver.quit()
```

### Testing


```python
output_file = r"C:\PT ESRI Indonesia\Project\P9 (Telkomsel - POI)\output\kemenperin_6.csv"
driver = webdriver.Chrome()
base_url = 'https://www.kemenperin.go.id/direktori-perusahaan?what=&prov=&hal='
#driver.get('https://www.kemenperin.go.id/direktori-perusahaan?what=&prov=&hal=1')

xpath_poiname = '//*[@id="newspaper-a"]/tbody/tr/td[2]/b'
xpath_address = '//*[@id="newspaper-a"]/tbody/tr/td[2]'

try :    
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile :
        fieldname = ['poi_name', 'address']
        writer = csv.DictWriter(csvfile, fieldnames=fieldname, delimiter=";")
            
        writer.writeheader()
            
        for page in range(3352,3578) :
            
            try :
                url_fix = base_url+str(page)
                driver.get(url_fix)
                time.sleep(2)
                
                try :
                    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, xpath_poiname)))
                    
                    element_poiname = driver.find_elements("xpath", xpath_poiname)
                    element_address = driver.find_elements("xpath", xpath_address)
                    #text_poiname = element_poiname.text
                    for poiname, addres in zip(element_poiname, element_address) :   
                        text_poiname = poiname.text.strip()
                        text_address = addres.text.strip() if addres else ""
                        writer.writerow({'poi_name': text_poiname, 'address' : text_address})
                        print(f"Page ke {page} - poi_name : {text_poiname} /// address : {text_address}")
                except Exception as e :
                    print(f"Error ni halaman {page} : {e}")
            except TimeoutError :
                print(f'Page nomor {page} bermasalah')
            
finally :
    driver.quit
```
