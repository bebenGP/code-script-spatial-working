```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Inisialisasi driver Selenium (misalnya, Chrome)
driver = webdriver.Chrome()

# Buka halaman web yang berisi tabel
driver.get("https://referensi.data.kemdikbud.go.id/pendidikan/paud")

# Mencari semua baris (tr) dalam tabel
rows = driver.find_elements(By.XPATH, '//*[@id="table1"]/tbody/tr')

# Data yang akan ditulis pada csv
data_provinsi = []

# Melewati baris-baris untuk mengklik nama provinsi dan mengakses halaman terkait
for row in range(0, len(rows)):  # Mulai dari indeks 0 untuk mengakses semua baris
    row_element = driver.find_elements(By.XPATH, '//*[@id="table1"]/tbody/tr')[row]  # Cari ulang baris
    columns = row_element.find_elements(By.TAG_NAME, 'td')
    if len(columns) >= 2:  # Pastikan terdapat cukup kolom
        nama_provinsi = columns[1].text.strip()

        # Tambahkan data provinsi ke list
        data_provinsi.append({
            "Nama Provinsi": nama_provinsi,
            "Kabupaten/Kota": []  # Inisialisasi list untuk data kabupaten/kota
        })

        # Cari tautan (link) dalam kolom nama provinsi dan klik
        link_province = columns[1].find_element(By.TAG_NAME, 'a')
        link_province.click()

        # Menambahkan penundaan sebelum mencoba mengakses tautan kabupaten/kota
        time.sleep(1)  # Tunggu 1 detik

        # Sekarang kita berada di halaman kabupaten/kota
        # Mencari semua baris (tr) dalam tabel
        city_rows = driver.find_elements(By.XPATH, '//*[@id="table1"]/tbody/tr')

        for city_row in range(0, len(city_rows)):  
            city_row_elements = driver.find_elements(By.XPATH, '//*[@id="table1"]/tbody/tr')
            if city_row < len(city_row_elements):
                city_row_element = city_row_elements[city_row]
                city_columns = city_row_element.find_elements(By.TAG_NAME, 'td')
                if len(city_columns) >= 2:
                    nama_kabupaten = city_columns[1].text.strip()

                    # Tambahkan data kabupaten/kota ke list provinsi saat ini
                    data_provinsi[-1]["Kabupaten/Kota"].append({
                        "Nama Kabupaten/Kota": nama_kabupaten,
                        "Kecamatan": []  
                    })

                    # Cari tautan (link) dalam kolom nama kabupaten/kota dan klik
                    link_city = city_columns[1].find_element(By.TAG_NAME, 'a')
                    link_city.click()

                    # Menambahkan penundaan sebelum mencoba mengakses tautan kecamatan
                    time.sleep(1)

                    # Sekarang kita berada di halaman kecamatan
                    # Mencari semua baris (tr) dalam tabel
                    kecamatan_rows = driver.find_elements(By.XPATH, '//*[@id="table1"]/tbody/tr')

                    for kecamatan_row in range(0, len(kecamatan_rows)):  
                        kecamatan_row_elements = driver.find_elements(By.XPATH, '//*[@id="table1"]/tbody/tr')
                        if kecamatan_row < len(kecamatan_row_elements):
                            kecamatan_row_element = kecamatan_row_elements[kecamatan_row]
                            kecamatan_columns = kecamatan_row_element.find_elements(By.TAG_NAME, 'td')
                            if len(kecamatan_columns) >= 2:
                                nama_kecamatan = kecamatan_columns[1].text.strip()

                                # Tambahkan data kecamatan ke list kabupaten/kota saat ini
                                data_provinsi[-1]["Kabupaten/Kota"][-1]["Kecamatan"].append({
                                    "Nama Kecamatan": nama_kecamatan,
                                    "Sekolah": []  
                                })

                                # Kembali ke halaman kabupaten/kota
                                driver.back()

                    # Kembali ke halaman provinsi
                    driver.back()

```
