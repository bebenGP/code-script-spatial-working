```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Inisialisasi driver Selenium (misalnya, Chrome)
driver = webdriver.Chrome()

# Buka halaman web yang berisi tabel
driver.get("https://referensi.data.kemdikbud.go.id/pendidikan/paud")

# Mencari semua baris (tr) dalam tabel
rows = driver.find_elements(By.XPATH, '//*[@id="table1"]/tbody/tr')

# Data yang akan ditulis pada csv 
data_provinsi = []

# Melewati baris-baris untuk mengklik nama provinsi dan mengakses halaman terkait
for row in range(1, len(rows)):  # Mulai dari indeks 1 untuk menghindari header
    row_element = driver.find_elements(By.XPATH, '//*[@id="table1"]/tbody/tr')[row]  # Cari ulang baris
    columns = row_element.find_elements(By.TAG_NAME, 'td')
    if len(columns) >= 2:  # Pastikan terdapat cukup kolom
        nama_provinsi = columns[1].text.strip()
        
         # Tambahkan data ke list
        data_provinsi.append({
            "Nama Provinsi": nama_provinsi,
        })
       
        # Cari tautan (link) dalam kolom nama provinsi dan klik
        link_province = columns[1].find_element(By.TAG_NAME, 'a')
        link_province.click()
        
        # Sekarang Anda berada di halaman provinsi
        # Anda dapat mengekstrak informasi lainnya dari halaman ini
        
        # Mencari tautan kabupaten/kota di halaman provinsi
        kabupaten_links = driver.find_elements(By.XPATH, '//a[contains(@href, "/pendidikan/paud/") and not(contains(@href, "/1"))]')
        kabupaten_data = []
        for kabupaten_link in kabupaten_links:
            nama_kabupaten = kabupaten_link.text.strip()
            
            # Tambahkan data kabupaten ke list
            kabupaten_data.append({
                "Kabupaten/Kota": nama_kabupaten,
            })
            
            # Klik tautan kabupaten untuk mengakses halaman kabupaten
            kabupaten_link.click()
            
            # Sekarang Anda berada di halaman kabupaten/kota
            # Anda dapat mengekstrak informasi lebih lanjut dari halaman ini
            
            # Kembali ke halaman provinsi
            driver.back()
        
        # Tambahkan data kabupaten/kota ke data provinsi
        data_provinsi[-1]["Kabupaten/Kota"] = kabupaten_data
        
        # Kembali ke halaman daftar provinsi
        driver.back()

# Cek data
for provinsi in data_provinsi:
    print(provinsi)

# Tutup browser saat selesai
driver.quit()
```
