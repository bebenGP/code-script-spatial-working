```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
file_data = r"C:\PT ESRI Indonesia\Project\P2 (Data Services)\Scraping Data\Raw\public_recreation\public_recreation.csv"
data = pd.read_csv(file_data, delimiter=';')
kolom = 'poi_name'
all_text = ' '.join(data[kolom].dropna())

# Membuat WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Tampilkan WordCloud menggunakan Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from collections import Counter

# Menghitung frekuensi kata-kata dalam teks
word_freq = Counter(all_text.split())

# Menampilkan kata-kata yang paling banyak muncul
most_common_words = word_freq.most_common(20)  # Mengambil 10 kata teratas

for word, freq in most_common_words:
    print(f"{word}: {freq}")
```
