```python
import requests
import csv
import time
import re
from bs4 import BeautifulSoup
```

```python
base_url = "https://indahonline.com/company/cabang/list/"
output_directory = r"C:\PT ESRI Indonesia\Project\P2 (Data Services)\Scraping Data\Raw\ekspedisi"
csv_file_path = f"{output_directory}\\indah_cargo_testing1.csv"
```

```python
with open(csv_file_path,'w', newline='', encoding='utf-8') as csvfile :
    
    # Bikin Field/Header
    fieldname = ['data']
    
    writer = csv.DictWriter(csvfile, fieldnames=fieldname , delimiter=',')
    writer.writeheader()
    
    for id_number in range (1, 1873) :
        print(id_number)
        
        time.sleep(3)
        
        # Hit request URL
        url = base_url + str(id_number)
        response = requests.get(url)
        
        # Cheking Status
        if response.status_code == 200 :
            # ambil response text
            data = response.text
            print(data)
            # Write csv di field/header 'data' berdasarkan response
            writer.writerow({'data': data})          
        else :
            print(f'Data tidak tersedia, nomor : {id_number}')
            
print("Scrapping Finish")
```
