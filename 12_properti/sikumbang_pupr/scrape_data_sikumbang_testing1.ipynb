```python
import requests
import csv
import time
import re
from bs4 import BeautifulSoup
```

```python
base_url = "https://sikumbang.tapera.go.id/ajax/lokasi/search?sort=terbaru&page="
output_directory = r"C:\PT ESRI Indonesia\Research\code-script\12_properti\sikumbang_pupr\result"
csv_file_path = f"{output_directory}\\sikumbang_tapera_testing1.csv"

csv_header = ["idLokasi", "namaPerumahan", "jenisPerumahan", "jumlahUnit", "jumlahUnitKomersil", "koordinatPerumahan", "status", "nama", 
              "harga", "kamarTidur", "kamarMandi", "jumlahLantai", "luasTanah", "luasBangunan", "spesifikasiAtap", "spesifikasiDinding",
              "spesifikasiLantai", "spesifikasiPondasi", "nama", "asosiasi", "pengembang", "foto", "tipeRumah", "kantorPemasaran", "tipeRusun", "wilayah"]
```

```python
with open(csv_file_path,'w', newline='', encoding='utf-8') as csvfile :
    
    writer = csv.DictWriter(csvfile, fieldnames=csv_header , delimiter=';')
    writer.writeheader()
    
    for id_number in range (1, 10) :
        print(id_number)
        
        time.sleep(2)
        
        # Hit request URL
        url = base_url + str(id_number)
        response = requests.get(url)
        print(response)

        # Cheking Status
        if response.status_code == 200 :
            # ambil response text
            data_list = response.json().get("data")
            print(data_list)
            for data in data_list :
                # Write csv di field/header 'data' berdasarkan response
                writer.writerow(data)
        else :
            print(f'Data tidak tersedia, nomor : {id_number}')
            
print("Scrapping Finish")
```
